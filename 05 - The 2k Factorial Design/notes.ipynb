### Analysis Procedure for a $2^k$ Design

1. Estimate factor effects
2. Form initial model
3. If the design is replicated, fit the full model
4. If there is no replication, form the model using a normal probability plot of the effects
5. Perform statistical testing
6. Refine model
7. Analyze residuals
8. Interpret results

### Sparsity of Effects Principle

As the number of factors increases, the total number of treatment combination increases which may result in a problem when resources are limited and the number of replicates the experimenter can employ is restricted.

To resolve this, the **unreplicated $2^k$ design** (also known as the single-replicate strategy) is used. However, with only one replicate, there is no internal estimation of error, hence we have to assume that certain high-order interactions are negligible and combine their mean squares to estimate the error.

This is known as the sparsity of effects principle. It removes the illusion that more variables inside a statistical model produces a greater model predictive accuracy, which in actual reality, produces over-fitting. The best performing model is somewhere between not too few variables and not too many variables.

Examine a normal probability plot of the estimates of the effects, and the effects that are negligible are normally distributed, with mean zero and constant variance and will tend to fall along a straight line on this plot, whereas significant effects will have non-zero means and will not lie along the straight line.

## The Chemical Catalyst Experiment

This experiment aims to give a sense of a **$2^2$ design**.

Consider an investigation into the effect of the concentration of the reactant and the amount of the catalyst on the conversion (yield) in a chemical process. 

The objective of the experiment was to determine if adjustments to either of these `2` variables would increase the yield. 

Let the reactant concentration be variable **A** and assume the `2`-levels of interest be `15` and `25`%. 

The catalyst is variable **B**, with the high-level denoting the use of `2` pounds of the catalyst and the low-level denoting the use of only `1` pound. 

The experiment is replicated `3` times, so there are `12` runs (`3 * 2^2`). The order in which the runs are made is random, so this is a completely randomised experiment.

Firstly, import the dataset

```r
dataset <- read.csv('chem.csv')
```

Display the table to get a sense of the data

```r
library(dplyr)
```

```r
dataset_tbl <- dataset %>% group_by(A, B) %>% summarise(Yields=paste0(Y, collapse=" , "), Total=sum(Y)) %>% arrange(desc(A), desc(B))
dataset_tbl$Var <- c("1", "b", "a", "ab")
dataset_tbl
```

From above, the treatment combinations effect are as follows (this is referred to as standard order (or [Yates's order](https://en.wikipedia.org/wiki/Yates_analysis)). Using this, we can see the contrast coefficients used in estimating the effects,

```r
effects <- data.frame(Effect=c("A", "B", "AB"),
                      Trt.1 =c( -1,  -1,    1),
                      Trt.a =c(  1,  -1,   -1),
                      Trt.b =c( -1,   1,   -1),
                      Trt.ab=c(  1,   1,    1))

effects$Avg.Total <- (1/(2^(2-1) * 3)) * ((dataset_tbl %>% filter(Var == "1"))$Total * effects$Trt.1 + 
                                          (dataset_tbl %>% filter(Var == "a"))$Total * effects$Trt.a + 
                                          (dataset_tbl %>% filter(Var == "b"))$Total * effects$Trt.b + 
                                          (dataset_tbl %>% filter(Var == "ab"))$Total * effects$Trt.ab)
effects
```

Visualize the clustering,

```r
install.packages('ggplot2')
```

```r
library(ggplot2)
```

```r
displayEffectsGraph <- function (effects) {
    effects.sd <- sd(effects$Avg.Total)
    effects.mean <- mean(effects$Avg.Total) 

    effects.prob <- data.frame(labels=effects$Effect, avg=effects$Avg.Total)
    effects.prob$zscore <- (effects$Avg.Total - effects.mean) / effects.sd
    effects.prob$prob <- pnorm(effects.prob$zscore, mean=0, sd=1, lower.tail=TRUE)

    ggplot(aes(x=avg, y=prob, label=labels), data=effects.prob) + 
        geom_point() + 
        geom_label() + 
        geom_smooth(method='lm', se=FALSE)
}
```

```r
displayEffectsGraph(effects)
```

Notice that **A** and **B** is far from the center. Using these variables in ANOVA,

```r
anova <- aov(Y ~ A + B, data=dataset)
summary(anova)
```

We can see that the p-values is less than `0.05` (alpha level) showing significant evidence of affect of **A** and **B**.

```r
with(dataset, interaction.plot(A, B, Y, type="b", pch=19, fixed=T, xlab="A",ylab="Yield"))
```

> The sparsity of effects principle in experiment is analogous to the parsimony principle in Statistical models. In general, one can choose to include more and more variables inside a statistical model and we get the illusion of greater model predictive accuracy. However, this is an illusion and we will be disappointed with the model performance when we apply it to predict future cases. This is due to over-fitting. The best performing model is somewhere between not too few variables and not too many variables. 

Suppose a chemical product is produced in a pressure vessel. A factorial experiment is carried out in the pilot plant to study the variables that may influence the filtration rate of this product. The four variables are temperature (**A**), pressure (**B**), concentration of formaldehyde (**C**), and stirring rate (**D**). Each variable is present at `2`-levels.

## The Filtration Rate Experiment (Extended)



Using the set dataset, the grand mean $\bar{y}$ we obtain is,

```r
mean(dataset$Y)
```

From the centerpoints, we obtain another `4` samples, where $x_1 = x_2 = x_3 = x_4 = 0$,

```r
points <- c(73,75,66,69)
```

```r
mean(points)
```

We see the difference as,

```r
diff <- mean(dataset$Y) - mean(points)
diff
```

The formula for the sums of pure quadratic (curvature), 

$SS_{PureQuadratic} = \large{\frac{n_fn_c(\bar{y}_c - \bar{y}_f)^2}{n_f + n_c}}$ 

```r
SSPQ <- (16 * 4 * (diff)**2) / (16 + 4)
SSPQ
```

Calculating the $MS_E$ gives,

$MS_E = \large{\frac{\sum_{i=1}^{n_c} (y_i - \bar{y}_c)^2}{n_c - 1}}$ 

```r
MSE <- sum((points - mean(points))**2) / (length(points) - 1)
MSE
```

The $F$-score gives,

```r
F <- SSPQ/MSE
F
```

The degrees of freedom is,

```r
df <- 4 - 1
df
```

```r
pf(F, df1=1, df2=df, lower.tail=FALSE)
```

The hypothesis is as follows,

$H_0: \sum_{j=1}^{k} \beta_{jj} = 0$

$H_1: \sum_{j=1}^{k} \beta_{jj} \neq 0$

Based on the p-value, given that it is less than `0.05`, there is no evidence of second-order curvature in the response over the region of exploration, hence we accept the null hypothesis.
