# The Randomised Completed Block Design (RCBD)

Suppose we have, in general, $a$ treatments that are to be compared and $b$ blocks, and there is one observation per treatment and block, and the order in which the treatments are run within each block is randomly determined.

<img src="Blocks.png" width="250" height="auto" />

(Source: [Montgomery, D. C., Custer, L. & McCarville, D. R., 2013, \"Design and analysis of experiments: Student solutions manual (8th ed.).\", Hoboken, NJ: John Wiley & Sons](https://www.amazon.com/Student-Solutions-Manual-Analysis-Experiments/dp/1118388194))

$y_{ij} = \mu + \tau_i + \beta_j + \epsilon_{ij}\ \ \forall i \in [1,a], j \in [1,b]$

The ANOVA equation becomes,

$SS_T = SS_{Treatments} + SS_{Blocks} + SS_E$

- $SS_{Treatments}$ have $a - 1$ degree of freedoms.
- $SS_{Blocks}$ have $b - 1$ degree of freedoms.
- $SS_T$ have $N - 1$ (where $N = a \times b$) degree of freedoms.

<img src="ANOVA.png" width="550" height="auto" />

(Source: [Montgomery, D. C., Custer, L. & McCarville, D. R., 2013, \"Design and analysis of experiments: Student solutions manual (8th ed.).\", Hoboken, NJ: John Wiley & Sons](https://www.amazon.com/Student-Solutions-Manual-Analysis-Experiments/dp/1118388194))

By isolating the effects of blocks, it can precisely model for the effects of treatments in the presence of known nuisance variables. 

Note that we only want to draw hypotheses and inferrence of the treatment factor, even though the blocks are considerably a factor, and in some case a two-way structure.

# The Latin Square Design

## Multi-Blocking Principle

Suppose that there are $c$ formulations to each of the $a$ treatments and $b$ blocks (where $a = b = c$), then we can use the Latin Square Design to eliminate `2` nuisance sources of variability, by systematically allowing blocking in `2` directions. Thus, the rows and columns represents `2` restrictions on randomisation. 

In general, a Latin square for $p$ variables (e.g. $p \times p$ Latin square) is a square containing $p$ rows and  columns, and each of the resulting $p^2$ cells contains one of the $p$ letters that corresponds to the treatments and each letter occurs once and only once in each row and column.

<img src="LatinSquares.png" width="300" height="auto" />

(Source: [Montgomery, D. C., Custer, L. & McCarville, D. R., 2013, \"Design and analysis of experiments: Student solutions manual (8th ed.).\", Hoboken, NJ: John Wiley & Sons](https://www.amazon.com/Student-Solutions-Manual-Analysis-Experiments/dp/1118388194))

$y_{ijk} = \mu + \alpha_i + \tau_j + \beta_k + \epsilon_{ijk}\ \ \forall i \in [1,p], j \in [1,p], k \in [1,p]$

- $y_{ijk}$ is the observation in the $i$-th row $k$-th column for the $j$-th treatment.
- $\mu$ is the overall mean.
- $\alpha_i$ is the $i$-th row effect.
- $\tau_j$ is the $j$-th treatment effect.
- $\beta_k$ is the $k$-th column effect.
- $\epsilon_{ijk}$ is the random error.

This is an effects model and is completely additive (meaning no interaction between rows, columns and treatments).

$SS_T = SS_{Rows} + SS_{Columns} + SS_{Treatments} + SS_E$

- $N = p^2$
- For each $SS$ have respective degrees of freedom $p^2 – 1 = (p – 1) + (p – 1) + (p – 1) + (p – 2)(p – 1)$
- $\epsilon_{ijk} \sim N(0, \sigma^2)$ where it is an independently chi-square random variable.

<img src="LatinSquareANOVA.png" width="550" height="auto" />

(Source: [Montgomery, D. C., Custer, L. & McCarville, D. R., 2013, \"Design and analysis of experiments: Student solutions manual (8th ed.).\", Hoboken, NJ: John Wiley & Sons](https://www.amazon.com/Student-Solutions-Manual-Analysis-Experiments/dp/1118388194))

## Graeco-Latin Square Design

Consider a $p \times p$ Latin square and superimpose on it with a second $p \times p$ Latin square in which the treatments are denoted by Greek letters.

<img src="GraecoLatinSquare.png" width="250" height="auto" />

(Source: [Montgomery, D. C., Custer, L. & McCarville, D. R., 2013, \"Design and analysis of experiments: Student solutions manual (8th ed.).\", Hoboken, NJ: John Wiley & Sons](https://www.amazon.com/Student-Solutions-Manual-Analysis-Experiments/dp/1118388194))

If the `2` squares when superimposed have the property that each Greek letter appears once and only once with each Latin letter, the two Latin squares are said to be orthogonal and the design obtained is called a Graeco-Latin square.

## Balanced Incomplete Block Design (BIBD)

In some experiment using Randomised block designs, we may not be able to run all the treatment combinations within each block (e.g. shortages of experimental apparatus or facilities).

The BIBD ensures that the treatment combinations used in each block would be selected in a balanced manner so that any pair of treatments occurs together the same number of times as any other pair, given that all treatment comparisons are equally important.

There are $a$ treatments and $b$ blocks, and each block can hold exactly $k$ (where $k < a$) treatments. Each block contains $k$ treatments, and each treatment occurs (or is replicated) $r$ times in the design, and there are $N = ar = bk$ total observations.

It may be constructed by taking $C_k^a$ blocks. and assigning a different combination of treatments to each block, but frequently obtained with fewer than $C_k^a$ blocks.

$y_{ij} = \mu + \tau_i + \beta_j + \epsilon_{ij}$

$SS_T = SS_{Treatments(adjusted)} + SS_{Blocks} + SS_E$

## The Vascular Graft Experiment

This experiment aims to give a sense of **Randomised Complete Block Design (RCBD)**. 

Suppose a medical device manufacturer would like to determine the levels of extrusion pressure (treatments), as some tubes in a production contain small and hard protrusions on the external surface which is a cause for rejection of the product (known as "flicks").

There may be significant batch-to-batch variation (`6` batches) due to manufacturing variation at the resin supplier side and natural variation in the material despite produce process of consistency with respect to parameters such as molecular weight, mean particle size, retention.

Hence, the product developer decides to investigate the effect of `4` different levels of extrusion pressure on the "flicks" via RCBD considering different batches of resin as blocks. 

Note that the order in which the extrusion pressures are tested within each block is random. 

The response variable is yield, or the percentage of tubes in the production run that does not contain "flicks".

Firstly, import the data

```r
dataset <- read.csv("vascular.csv")
```

Display table to get a sense of the data

```r
library(dplyr)
```

```r
dataset_tbl <- data.frame(Treatment=unique(dataset$Treatment),
                          Block.1=(dataset %>% filter(Block == "batch1"))$Yield,
                          Block.2=(dataset %>% filter(Block == "batch2"))$Yield,
                          Block.3=(dataset %>% filter(Block == "batch3"))$Yield,
                          Block.4=(dataset %>% filter(Block == "batch4"))$Yield,
                          Block.5=(dataset %>% filter(Block == "batch5"))$Yield,
                          Block.6=(dataset %>% filter(Block == "batch6"))$Yield,
                          Totals=(dataset %>% group_by(Treatment) %>% summarise(Totals=sum(Yield)))$Totals)
dataset_tbl
```

Display totals of each blocks

```r
data.frame(Block.1=sum(dataset_tbl$Block.1),
           Block.2=sum(dataset_tbl$Block.2),
           Block.3=sum(dataset_tbl$Block.3),
           Block.4=sum(dataset_tbl$Block.4),
           Block.5=sum(dataset_tbl$Block.5),
           Block.6=sum(dataset_tbl$Block.6),
           Totals=sum(dataset_tbl$Totals))
```

Display boxplot for data

```r
boxplot(Yield ~ Treatment, data=dataset, 
        main="Comparative Boxplot",
        xlab="Treatment",
        ylab="% Not Containing \"flicks\"")
```

```r
boxplot(Yield ~ Block, data=dataset, 
        main="Comparative Boxplot",
        xlab="Batch",
        ylab="% Not Containing \"flicks\"")
```

Using blocking in ANOVA

```r
anova <- aov(Yield ~ Treatment + Block, data=dataset)
summary(anova)
```

Display the Residual Diagnostic Plots

```r
par(mfrow=c(2,2))
plot(anova)
```

Set the factor as the treatment level as it is the controllable variable,

```r
dataset$Treatment <- as.factor(dataset$Treatment)
```

Using no blocking in ANOVA

Notice that although the end-result is still less than `0.05`, it still risks of ending up with a wrong conclusion as there is an increase of `0.0234/0.00192 = 12` times.

```r
anova <- aov(Yield ~ Treatment, data=dataset)
summary(anova)
```

Display the Residual Diagnostic Plots

```r
par(mfrow=c(2,2))
plot(anova)
```

Hypothesis: $H_0: \tau_1 = \tau_2 = ... = \tau_a = 0, H_1: \tau_i \neq 0\ \ \exists i$. Conclusion: 

We reject the null hypothesis as p-value < `0.05` (alpha level), noting evidence that the extrusion pressure significantly affects the mean yield (_`%` of tubes in the production run not containing "flicks"_).

(In this case, the treatment is the level of PSIs, and we find significant variation in the % of good yield indicating that there is a signifcant amount of pressure affecting the mean yield and subsequently a rejection of the batches.)

## The Rocket Propellant Experiment

This experiment aims to give a sense of **Latin Square Design**.

Suppose that an experimenter is studying the effects of `5` different formulations ("A", ..., "E") of a rocket propellant used in aircrew escape systems on the observed burning rate, and each is mixed from a batch of raw material that is only large enough for `5` formulations to be tested.

The formulations are prepared by several operators, and there may be substantial differences in the skills and experience of the operators. 

This gives `2` nuisance variables in the design: batches of raw material and operators. 

We would need to test each formulation exactly once in each batch of raw material and for each formulation to be prepared exactly once by each of five operators. Therefore, we use the LSD to do such.

Firstly, import the data

```r
dataset <- read.csv("rocket1.csv")
```

Display table to get a sense of the data

```r
data.frame(Batch=unique(dataset$Batch),
           Operator.1=(dataset %>% filter(Operator == "op1") %>% summarise(Pair=paste(Treatment, " = ", Burn.Rate)))$Pair,
           Operator.2=(dataset %>% filter(Operator == "op2") %>% summarise(Pair=paste(Treatment, " = ", Burn.Rate)))$Pair, 
           Operator.3=(dataset %>% filter(Operator == "op3") %>% summarise(Pair=paste(Treatment, " = ", Burn.Rate)))$Pair, 
           Operator.4=(dataset %>% filter(Operator == "op4") %>% summarise(Pair=paste(Treatment, " = ", Burn.Rate)))$Pair, 
           Operator.5=(dataset %>% filter(Operator == "op5") %>% summarise(Pair=paste(Treatment, " = ", Burn.Rate)))$Pair,
           Totals=(dataset %>% group_by(Batch) %>% summarise(Totals=sum(Burn.Rate)))$Totals)
```

There is evidence of significant variation in operators used due to its p-value < `0.05` (alpha level).

There is no evidence of significant variation in the batches due to its p-value > `0.05` (alpha level).

(In this case, there is variation in the effects of `5` different formulations which affects the burning rate.)

Taking the batch 3 row,

| Batch | Operator.1 | Operator.2 | Operator.3 | Operator.4 | Operator.5 |
| -- | -- | -- | -- | -- | -- |
| batch3 | C=`18` | D=`38` | E=`26` | A=`27` | B=`21` |  

To illustrate,

$y_{ijk} = \mu + \alpha_i + \tau_j + \beta_k + \epsilon_{ijk}$

```r
# Take batch1 for example,
# C=18	D=38	E=26	A=27	B=21

batch <- data.frame(Burn.Rate=c(18, 38, 26, 27, 21))
batch$Mean <- rep(mean(dataset$Burn.Rate), times=nrow(batch))
batch$Treatment.Effect <- c(
    mean((dataset %>% filter(Treatment == "C"))$Burn.Rate) - mean(dataset$Burn.Rate),
    mean((dataset %>% filter(Treatment == "D"))$Burn.Rate) - mean(dataset$Burn.Rate),
    mean((dataset %>% filter(Treatment == "E"))$Burn.Rate) - mean(dataset$Burn.Rate),
    mean((dataset %>% filter(Treatment == "A"))$Burn.Rate) - mean(dataset$Burn.Rate),
    mean((dataset %>% filter(Treatment == "B"))$Burn.Rate) - mean(dataset$Burn.Rate)
)
batch$Row.Effect <- rep(mean(batch$Burn.Rate), times=nrow(batch))
batch$Col.Effect <- c(
    mean((dataset %>% filter(Operator == "op1"))$Burn.Rate) - mean(dataset$Burn.Rate),
    mean((dataset %>% filter(Operator == "op2"))$Burn.Rate) - mean(dataset$Burn.Rate),
    mean((dataset %>% filter(Operator == "op3"))$Burn.Rate) - mean(dataset$Burn.Rate),
    mean((dataset %>% filter(Operator == "op4"))$Burn.Rate) - mean(dataset$Burn.Rate),
    mean((dataset %>% filter(Operator == "op5"))$Burn.Rate) - mean(dataset$Burn.Rate)
)
batch$Error <- batch$Burn.Rate - batch$Mean - batch$Treatment.Effect - batch$Row.Effect - batch$Col.Effect
batch$Computed <- batch$Mean + batch$Treatment.Effect + batch$Row.Effect + batch$Col.Effect + batch$Error
batch
```

## The Chemical Catalyst Experiment

This experiment aims to give a sense of **Randomised Incomplete Block Design (RIBD)**.

Suppose that an engineer thinks that the time of reaction for a chemical process depends on the type out of the `4` catalyst used.

The experiment procedure consists of selecting a batch of raw material, loading the pilot plant, applying each catalyst in a separate run of the pilot plant, and observing the reaction time. 

Because variations in the batches of raw material may affect the performance of the catalysts, the engineer decides to use batches of raw material as blocks, and each batch is only large enough to permit `3` catalysts to be run, hence a RIBD is used. 

The **Balanced Incomplete Block Design (BIBD)** is shown to display how each block is randomised.

Firstly, import the data

```r
dataset <- read.csv("chemical.csv")
```

Display imbalanced table for visualisation. Notice that for each treatment, the batches are randomised.

```r
dataset %>% group_by(Treatment) %>% summarise(Batches=paste(Batch, " = ", Yield, collapse=" , "), Totals=sum(Yield))
```

Balance the table by setting `NA` for missing treatments to balance the design in batches.

```r
for (t in unique(dataset$Treatment)) {
    for (b in unique(dataset$Batch)) {
        if (nrow(dataset %>% filter(Batch == b, Treatment == t)) == 0) {
            dataset <- rbind(dataset, list(Treatment=t, Batch=b, Yield=NA), stringsAsFactors=FALSE)
        }
    }
}
```

```r
dataset <- dataset %>% arrange(Treatment, Batch)
```

Display balanced table for visualisation

```r
data.frame(Treatment=unique(dataset$Treatment),
           Batch.1=(dataset %>% filter(Batch == 1))$Yield,
           Batch.2=(dataset %>% filter(Batch == 2))$Yield,
           Batch.3=(dataset %>% filter(Batch == 3))$Yield,
           Batch.4=(dataset %>% filter(Batch == 4))$Yield,
           Totals=(dataset %>% group_by(Treatment) %>% summarise(Totals=sum(Yield, na.rm=TRUE)))$Totals)
```

Display the totals for batches

```r
data.frame(Batch.1=sum((dataset %>% filter(Batch == 1))$Yield, na.rm=TRUE),
           Batch.2=sum((dataset %>% filter(Batch == 2))$Yield, na.rm=TRUE),
           Batch.3=sum((dataset %>% filter(Batch == 3))$Yield, na.rm=TRUE),
           Batch.4=sum((dataset %>% filter(Batch == 4))$Yield, na.rm=TRUE),
           Totals=sum((dataset %>% group_by(Treatment) %>% summarise(Totals=sum(Yield, na.rm=TRUE)))$Totals))
```

Since batch is the factor that the engineer is controlling,

```r
dataset$Batch <- factor(dataset$Batch)
```

Using blocking in ANOVA

```r
# Independent extraction of the variation between Treatment and Batch
anova <- aov(Yield ~ Treatment + Batch + Error(Batch), data=dataset)
summary(anova)
```

Hypothesis: $H_0: \tau_1 = \tau_2 = ... = \tau_a = 0, H_1: \tau_i \neq 0\ \ \exists i$. Conclusion: 

We reject the null hypothesis as p-value < `0.05` (alpha level), noting evidence of significant variation (or difference) in the time reaction depending on the catalyst (treatment) used.

(In this case, we want to determine the time of reaction for a chemical process depending on the type of catalyst used.)
